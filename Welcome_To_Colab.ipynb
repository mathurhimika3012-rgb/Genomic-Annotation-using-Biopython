```
# Experiment 1: Basic Python + File Handling + Sequence operations

dna_seq = Seq("ATGCGTATCGTAGCTAGCTAGCATCG")

complement_seq = dna_seq.complement()
reverse_complement_seq = dna_seq.reverse_complement()
mrna_seq = dna_seq.transcribe()
protein_seq = dna_seq.translate(to_stop=True)

# Store in FASTA (optional)
rec = SeqRecord(dna_seq, id="example_seq", description="Sample DNA")
SeqIO.write(rec, "example_seq.fasta", "fasta")

# Display all requested sequences
pd.DataFrame({
    "Type": [
        "DNA Sequence",
        "Complement Sequence",
        "Reverse Complement Sequence",
        "mRNA Sequence",
        "Protein Sequence"
    ],
    "Sequence": [
        str(dna_seq),
        str(complement_seq),
        str(reverse_complement_seq),
        str(mrna_seq),
        str(protein_seq)
    ]
})
```

```
# Experiment 2: Getting started with Biopython  Fetch GenBank record

handle = Entrez.efetch(db="nucleotide", id=ACC, rettype="gb", retmode="text")
record = SeqIO.read(handle, "genbank")
handle.close()

pd.DataFrame({
    "Accession ID": [record.id],
    "Organism": [record.annotations.get("organism", "NA")],
    "Description": [record.description],
    "Sequence Length (bp)": [len(record.seq)]
})
```

```
# Experiment 3: GC content calculation + sequence writing

dna = record.seq
gc_percent = gc_fraction(dna) * 100

SeqIO.write(record, f"{ACC}.fasta", "fasta")

pd.DataFrame({
    "Sequence Length (bp)": [len(dna)],
    "GC Content (%)": [round(gc_percent, 2)],
    "FASTA File Saved?": ["Yes"]
})
```

```
# Experiment 4: Working with Sequence Alignment objects

seq1 = dna[:300]
seq2 = dna[20:320]

alignment = pairwise2.align.globalxx(seq1, seq2, one_alignment_only=True)[0]

msa = MultipleSeqAlignment([
    SeqRecord(Seq(alignment.seqA), id="Seq1"),
    SeqRecord(Seq(alignment.seqB), id="Seq2")
])

AlignIO.write(msa, "alignment.aln", "clustal")

pd.DataFrame({
    "Alignment Score": [alignment.score],
    "Aligned Length": [len(alignment.seqA)],
    "Alignment File (alignment.aln)": ["Saved"]
})
```

```
# Experiment 5: Running BLAST search using NCBIWWW.qblast

query_seq = dna[:400]  # Use first 400 bp for faster BLAST

result_handle = NCBIWWW.qblast("blastn", "nt", query_seq)
with open("blast_results.xml", "w") as f:
    f.write(result_handle.read())
result_handle.close()

blast_record = NCBIXML.read(open("blast_results.xml"))

hits = []
for aln in blast_record.alignments[:3]:   # Top 3 hits
    hsp = aln.hsps[0]
    hits.append({
        "Hit ID": aln.hit_id,
        "Definition": aln.hit_def,
        "Score": hsp.score,
        "E-value": hsp.expect
    })

pd.DataFrame(hits)
```

```
# Experiment 9: Phylogenetics with Bio.Phylo

from Bio import Phylo

# ==== STEP 1:====
newick_tree = "(((A,B),(C,D)),(E,F,G));"

# Save to file (Phylo.read() requires a file)
with open("simple.dnd", "w") as f:
    f.write(newick_tree)

# ==== STEP 2: Read the tree ====
tree = Phylo.read("simple.dnd", "newick")

print("NEWICK TREE STRUCTURE:")
print(tree)

# ==== STEP 3: Display ASCII tree ====
print("\nASCII TREE:")
Phylo.draw_ascii(tree)

# ==== STEP 4: Convert to phyloXML ====
tree_xml = tree.as_phyloxml()

from Bio.Phylo.PhyloXML import Phylogeny
phyloxml_tree = Phylogeny.from_tree(tree)

# ==== STEP 5: Color nodes (E & F) ====
# Follows the manual style:
tree.root.color = (128, 128, 128)  # grey root color

# Find most recent common ancestor of E and F:
mrca = tree.common_ancestor({"name": "E"}, {"name": "F"})
mrca.color = "salmon"

# Color another clade (for example: index 0,1)
try:
    tree.clade[0].clade[1].color = "blue"
except:
    pass  # If tree shape differs

# ==== STEP 6: Draw the final colored tree ====
print("\nFINAL TREE (color info shown in text):")
Phylo.draw(tree)   # Graphical tree (Colab will show image)
```

```
# EXPERIMENT 10: Simple Motif Search in NM_001369735
# Himika Mathur

from Bio import Entrez, SeqIO
import matplotlib.pyplot as plt

Entrez.email = "mathurhimika3012@gmail.com"   # required by NCBI

# Step 1: Fetch the nucleotide sequence
handle = Entrez.efetch(db="nucleotide", id="NM_001369735", rettype="fasta", retmode="text")
record = SeqIO.read(handle, "fasta")
handle.close()

sequence = str(record.seq)
print("Sequence length:", len(sequence))

# Step 2: Define motif to search
motif = "ATG"     # start codon (you can change to any motif)
motif_len = len(motif)

print(f"\nSearching for motif: {motif}\n")

# Step 3: Search for motif positions
positions = []

for i in range(len(sequence) - motif_len):
    if sequence[i:i+motif_len] == motif:
        positions.append(i)

# Step 4: Output results
if positions:
    print("Motif found at positions (first 25 shown):")
    print(positions[:25], "...")    # show first 25 positions
    print("Total occurrences:", len(positions))
else:
    print("Motif not found in this sequence.")

# Step 5: Plot distribution (simple CA-1 style)
if positions:
    plt.hist(positions, bins=20, edgecolor='black')
    plt.title(f"Distribution of motif '{motif}' in NM_001369735")
    plt.xlabel("Sequence position")
    plt.ylabel("Frequency")
    plt.show()
```

```
# EXPERIMENT 10  Motif Search (Offline Version using uploaded FASTA)
# Himika Mathur

from Bio import SeqIO
import matplotlib.pyplot as plt
import re # Import regular expression module

# Step 1: Load the FASTA file
with open("/NM_001369735.1 Homo sapiens paterna.txt", "r", encoding="latin-1") as handle:
    record = SeqIO.read(handle, "fasta")

# Explicitly decode the sequence data if it contains non-ASCII characters
# record.seq._data is the internal bytes representation
try:
    # Try decoding internal bytes as ASCII first (standard for Biopython)
    sequence = record.seq._data.decode("ascii")
except UnicodeDecodeError:
    # If ASCII decoding fails, it means the sequence bytes contained non-ASCII.
    # We should decode it with the same encoding used for the file.
    sequence = record.seq._data.decode("latin-1")

# Now, clean the sequence string to ensure only valid nucleotide characters remain
# Remove any characters that are not A, T, C, G, N (case-insensitive)
sequence = re.sub(r'[^ATCGN]', '', sequence, flags=re.IGNORECASE)
sequence = sequence.upper() # Convert to uppercase for consistent searching

print("Sequence length after cleaning:", len(sequence))

# Step 2: Define motif
motif = "ATG"          # you can change this
motif_len = len(motif)

print(f"\nSearching for motif: {motif}\n")

# Step 3: Find motif positions
positions = []
for i in range(len(sequence) - motif_len):
    if sequence[i:i+motif_len] == motif:
        positions.append(i)

# Step 4: Print results
if positions:
    print("Motif found at positions:")
    print(positions[:25], " ...")   # show first 25
    print("Total occurrences:", len(positions))
else:
    print("Motif NOT found in sequence.")

# Step 5: Plot distribution
if positions:
    plt.hist(positions, bins=20, edgecolor='black')
    plt.title(f"Distribution of motif '{motif}' in NM_001369735")
    plt.xlabel("Sequence position")
    plt.ylabel("Frequency")
    plt.show()
```
